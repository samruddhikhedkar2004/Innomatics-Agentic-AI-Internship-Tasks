## Problem 1: Mobile Recharge Validation System

**Problem Statement**

Create a function to validate a mobile recharge request.

**Rules:**

- Recharge amount must be ≥ ₹50
- Valid plans: 199, 299, 399, 599
- User should be allowed to retry if invalid


```python

def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]
    if amount < 50:
        print("Recharge amount must be at least ₹50.")
        return False
    elif amount not in valid_plans:
        print("Invalid plan selected.")
        print("Available plans:", valid_plans)
        return False
    else:
        print("Recharge Successful!")
        return True

while True:
    try:
        amount = int(input("Enter recharge amount: ₹"))
        if validate_recharge(amount):
            break   # Exit loop if valid
        print("Please try again")
    except ValueError:
        print("Please enter a valid numeric amount")
```

## Problem 2: Inventory Reorder Alert System

**Problem Statement :** Given a dictionary of products and stock quantities:
- Print Reorder Alert if stock < 15
- Print Stock OK otherwise

```python
products = {
    'Mobile': 15,
    'Laptop': 30,
    'Headphones':8,
    'Smart TV': 5
}

# Create a function to check stock
def check_stock(product, quantity):
    if quantity < 15:
        print(f"{product}: Reorder Alert")
    else:
        print(f'{product}: Stock OK')

# Loop through inventory
for product, quantity in products.items():
    check_stock(product, quantity)
```

## Problem 3: Student Result Processing System

**Problem Statement :**
Create a function that:
- Accepts a list of student marks
- Calculates average marks
- Prints Pass if average ≥ 50 else Fail

```python
def calculate_result(marks):
    total = 0

    for mark in marks:
        total += mark

    average = total / len(marks)

    if average >= 50:
        result = 'Pass'
    else:
        result = "Fail"
    return average, result

# input
student_marks = [35, 48, 70, 85,90]
average, result = calculate_result(student_marks)
print('Average Marks:', average)
print('Result:', result)
```

## Problem 4: Cab Fare Estimator with Retry Option

**Problem Statement :**
Create a cab fare calculator:

- Base fare = ₹50
- ₹12 per km
- Peak hour → 25% extra
- Allow retry using while loop


```python
# Function to calculate fare
def calculate_fare(distance, peak):
    base_fare = 50
    per_km_rate = 12

    fare = base_fare + (distance* per_km_rate)

    if peak:
        fare *= 1.25  #adding extra 25%

    return fare

# Retry System
while True:
    distance = float(input('Enter distance in km: '))
    peak_hour = input('Is it peak hour? (yes/No): ').lower()

    peak = True if peak_hour == 'yes' else False

    total_fare = calculate_fare(distance, peak)

    print(f"Total Fare: {total_fare: .2f}")

    retry = input('Do you want to calculate again? (yes/No): ').lower()
    if retry != 'yes':
        break
```

## Problem 5: Employee Attendance Eligibility Checker

**Problem Statement :**
Create a function that:

- Accepts attendance list ("P", "A")
- Calculates attendance percentage
- Returns Eligible if ≥ 75% else Not Eligible


```python
def check_attendance(attendance):
    present_days = 0

    # Count no of P
    for day in attendance:
        if day == 'P':
            present_days +=1

    total_days = len(attendance)
    percentage = (present_days / total_days)*100

    if percentage >= 75:
        return percentage, 'Eligible'
    else:
        return percentage, 'Not Eligible'

attendance_list = ['P','P','A','P','P','A','P','P']

percentage, status = check_attendance(attendance_list)

print('Attendance Percentage:',percentage)
print('Eligibility Status:',status)
```

## Problem 6: Password Strength Checker

**Problem Statement :**
Create a function to check password strength.

- Minimum length = 8
- Must contain at least one digit
- Must contain at least one special character (@#$)

```python
def check_password(password):

    if len(password) < 8:
        return 'Weak Password'
    
    has_digit = False
    has_special = False

    for char in password:
        if char.isdigit():
            has_digit = True
        if char in "@#$":
            has_special = True

    if has_digit and has_special:
        return 'Strong Password'
    else:
        return 'Weak Password'

user_password = input('Enter your password: ')
result = check_password(user_password)
print(result)
```
