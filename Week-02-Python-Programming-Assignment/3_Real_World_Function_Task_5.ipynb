## Problem 3: Agriculture – Rainfall Adequacy Checker
**Problem Statement:**
Create a function that analyzes seasonal rainfall data.

**Conditions:**
- Calculate average rainfall
- Average ≥ required level → Adequate Rainfall
- Otherwise → Inadequate Rainfall


```python
def analyze_seasonal_rainfall(rainfall_data, required_level):
    total = 0

    # Calculate total rainfall using loop
    for rain in rainfall_data:
        total += rain

    average = total / len(rainfall_data)

    if average >= required_level:
        status = "Adequate rainfall"
    else:
        status = "Inadequate rainfall"

    return average, status

# Example usage
rainfall_data = [60,70,80,85]
required_level = 65

avg_rainfall, rainfall_status = analyze_seasonal_rainfall(rainfall_data, required_level)

print('Average Rainfall:', avg_rainfall)
print('Rainfall Status:', rainfall_status)
```

## Problem 4: Social Media – Duplicate Account Detection
**Problem Statement:**
Create a function to detect duplicate usernames in a system.

**Conditions:**
- If duplicates exist → Flag accounts
- Else → Accounts are unique


```python
# Function to check duplicate accounts
def check_duplicate_accounts(usernames):
    unique_users = set(usernames)

    if len(usernames) != len(unique_users):
        return 'Duplicate accounts found: yes'
    else:
        return 'Duplicate accounts found: no'

# Example usage
user_list = ['max', 'sara','max','bob']

result = check_duplicate_accounts(user_list)
print(result)
```

## Problem 6: Agriculture – Premium Crop Price Filter
**Problem Statement**
Create a function that filters crop prices above a premium threshold.

**Conditions:**
- Price > ₹2000 → Premium Crop
- Else → Regular Crop


```python
def filter_premium_crops(prices):
    premium_crops = []

    for price in prices:
        if price > 2000:
            premium_crops.append(price)
    
    return premium_crops

# Example usage
crop_prices = [1500, 2500, 3000, 1800]
premium_crops = filter_premium_crops(crop_prices)

print('Premium Crop Prices:', premium_crops)
```

## Problem 9: E-Learning – Student Attendance Eligibility System
**Problem Statement:**
Create a function to calculate attendance percentage and check exam eligibility.

**Conditions:**
- Attendance ≥ 75% → Eligible
- Attendance < 75% → Not Eligible


```python
def calculate_attendance_percentage(attendance):
    
    present_days = 0

    for day in attendance:
        if day == 'P':
            present_days += 1
    total_days = len(attendance)
    percentage = (present_days / total_days) * 100

    if percentage >= 75:
        status = 'Eligible'
    else:
        status = 'Not Eligible'
    
    return percentage, status

# Example usage
attendance_record = ['P', 'P', 'P', 'P', 'A', 'P', 'P']
attendance_percentage, eligibility_status = calculate_attendance_percentage(attendance_record)

print(f'Attendance Percentage: {attendance_percentage:.2f}%')
print('Eligibility Status:', eligibility_status)
```

## Problem 10: Smart Electricity Bill Analyzer
**Problem Statement:**
You are building a smart electricity billing system for residential users.
Create a function that calculates the electricity bill and classifies usage level based on units consumed.

**Billing Rules:**
- First 100 units → ₹3 per unit
- Next 100 units (101–200) → ₹5 per unit
- Above 200 units → ₹7 per unit

**Usage Classification:**
- Total bill < ₹500 → Low Usage
- ₹500 – ₹1500 → Moderate Usage
- ₹1500 → High Usage


```python
def calculate_bill(units):

    if units <= 100:
        bill_amount = units * 3
    elif units <= 200:
        bill_amount = (100 * 3) + ((units - 100) * 5)
    else:
        bill_amount = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    if bill_amount < 500:
        status = 'Low Usage'
    elif bill_amount <= 1500:
        status = 'Moderate Usage'
    else:
        status = 'High Usage'

    return bill_amount, status

# Example usage
units_consumed = 250
bill, usage_status = calculate_bill(units_consumed)

print('Units Consumed:', units_consumed)
print('Total Bill:', bill)
print('Usage Status:', usage_status) 
```
