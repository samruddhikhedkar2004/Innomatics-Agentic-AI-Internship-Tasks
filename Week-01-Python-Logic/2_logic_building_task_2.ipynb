## 1. Unique Words in a Sentence

**Problem Statement**
Given a sentence, identify all unique words and display:
The count of unique words
The unique words themselves

**Input**
Enter sentence: python is easy and python is powerful


```python
sentence = 'python is easy and python is powerful'

words = sentence.split()
unique_words = set(words)

print('Unique words count:', len(unique_words))
print('Unique words:', unique_words)
```

## 2. Highest Salary from Employee Data


**Problem Statement**

Store employee names and their salaries, and determine which employee has the highest salary.

**Input Format**

A dictionary where:
Keys represent employee names
Values represent employee salaries



```python
employees ={
    'Ravi': 75000,
    'Anita': 68000,
    'Kiran': 72000
}

highest_salary = 0
highest_employee = ''

for name, salary in employees.items():
    if salary > highest_salary:
        highest_salary = salary
        highest_employee = name

print('Highest Salary:', highest_employee, '-', highest_salary)
```

## 3. Find Maximum and Minimum Values

**Problem Statement**

Find the maximum and minimum values from a list of numbers.

**Input Format**

A list of integers where each element represents a numeric value.
Example Input
numbers = [45, 22, 89, 10, 66]


```python
numbers = [45,22,89,10,66]

max_value = numbers[0]
min_value = numbers[0]

for num in numbers:
    if num > max_value:
        max_value = num
    if num < min_value:
        min_value = num

print('List:', numbers)
print('Max:', max_value)
print('Min:', min_value)
```

## 4. Count Products Above a Price Threshold

**Problem Statement**

Count how many products have a price greater than 1000.

**Input Format**

A list of integers where each element represents the price of a product.
Example Input
prices = [450, 1200, 899, 1500, 300]



```python
prices = [450, 1200, 899, 1500, 300]

count = 0

for price in prices:
    if price > 1000:
        count +=1

print('Products above 1000:', count)
```

## 5. Calculate Attendance Percentage

**Problem Statement**

Calculate the attendance percentage from a list containing "P" (Present) and "A" (Absent).


```python
attendance = ['P','P','A','P','A','P']

present_count = 0

for day in attendance:
    if day == 'P':
        present_count +=1

percentage = (present_count/ len(attendance))*100
print('Attendance Percentage:', percentage)
```

## 7. Count Character Frequency

**Problem Statement**

Count how many times each character appears in a given string.


```python
word = 'Internship'
char_count = {}

#Loop through each letter in word to find the freq and increment by 1
for ch in word:
    if ch in char_count:
        char_count[ch] +=1
    else:
        char_count[ch] = 1

print(char_count)
```

## 10. Calculate Average Marks

**Problem Statement**

Calculate the average marks of a student from a list of marks.


```python
marks = [70,75,80,65]

total = 0

for mark in marks:
    total += mark

average = total / len(marks)

print('Average Marks:', average)
```
