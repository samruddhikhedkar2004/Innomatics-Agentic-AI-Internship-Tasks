**1. User Login Check**

```python
# User credentials
username = 'admin'
password = '1234'

# Check login
if username == 'admin' and password =='1234':
    print('Login Successful')
else:
    print('Invalid Credentials')
```

**2. Pass/ Fail Analyzer**

```python
marks = [45,78,90,33,60]
pass_count = 0
fail_count = 0

#analyse marks
for m in marks:
    if m>= 50:
        pass_count +=1
    else:
        fail_count +=1
print('Passed Students:', pass_count)
print('Failed Students:', fail_count)
```

**3. Simple Data Cleaner**

```python
names = ['Alice', 'bob', 'CHARLIE']

cleaned_names = []

for name in names:
    cleaned_names.append(name.strip().lower())  #convert all names to lowercase & remove extra spaces

print('Cleaned Names:', cleaned_names)
```

**4. Message Length Analyzer**

```python
messages = ['Hi','Welcome to the platform', 'OK']

for msg in messages:
    length = len(msg)
    print(f"Message:' '{msg}' | Length: {length}")

    if length > 10:
        print("Flag message - longer than 10 characters")
```

**5. Error Message Detector**

```python
logs = ['INFO', 'ERROR', 'WARNING', 'ERROR']

error_count = 0

# Count errors
for log in logs:
    if log == 'ERROR':
        error_count +=1

print('Total ERROR messages:', error_count)
```
